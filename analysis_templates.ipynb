# Analysis Templates
Load outputs and plot spectra.


```python
import numpy as np, matplotlib.pyplot as plt
# RF S-params (stub example)
try:
    sparams = np.loadtxt('../rf/out_rf/sparams_stub.csv', delimiter=',', skiprows=1)
    plt.figure(); plt.plot(sparams[:,0], sparams[:,1], label='|S11|'); plt.plot(sparams[:,0], sparams[:,2], label='|S21|');
    plt.xlabel('GHz'); plt.legend(); plt.title('RF S-parameters'); plt.show()
except Exception as e:
    print('RF data not found yet:', e)
```

```python
import numpy as np, matplotlib.pyplot as plt
# Roof spectrum (stub example)
try:
    spec = np.loadtxt('../roof/out_roof/spectrum_stub.csv', delimiter=',', skiprows=1)
    plt.figure(); plt.plot(spec[:,0], spec[:,1], label='Reflectance'); plt.plot(spec[:,0], spec[:,2], label='Emissivity');
    plt.xlabel('Wavelength (um)'); plt.legend(); plt.title('Roof Stack Spectrum'); plt.show()
except Exception as e:
    print('Roof data not found yet:', e)
```
